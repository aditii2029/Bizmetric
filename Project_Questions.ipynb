Write a Python program to generate a hotel bill .

The bill should display item name, quantity, price, and total amount.

Take input from the user.

```python

total = 0

n = int(input("How many items: "))

for i in range(n):
    name = input("Item name: ")
    qty = int(input("Quantity: "))
    price = int(input("Price: "))

    amt = qty * price
    total = total + amt

    print(name, "Amount =", amt)

print("Total Bill =", total)
```

Create a function to check age eligibility for given customer based on DOB. Function will take two input DOB and ELIGIBILITY age.

```python
def check_eligibility():
    try:
        dob_year = int(input("Enter birth year: "))
        eligibility_age = int(input("Enter eligibility age: "))

        current_year = 2026
        age = current_year - dob_year

        if age >= eligibility_age:
            print("Eligible")
        else:
            print("Not Eligible")

    except:
        print("Invalid input")


check_eligibility()
```
