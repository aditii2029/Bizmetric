```python
#11. Ask user to enter the marks and define if the user got distinction,
#first class, second class, pass, Fails
marks = int(input("Enter marks: "))

if marks > 80:
    print("You got distinction")
elif marks > 60:
    print("You got First class")
elif marks > 40:
    print("You got second class")
elif marks >= 35:
    print("PASS")
else:
    print("FAIL")
```

```python
#12. Ask user to enter information about salary of the employee per year and rating received as A, B, C, D
salary = float(input("Enter salary: "))
rating = input("Enter rating A/B/C/D: ")

if salary <= 5:
    inc = {'A':0.16,'B':0.12,'C':0.10,'D':0.06}
elif salary <= 10:
    inc = {'A':0.14,'B':0.10,'C':0.08,'D':0.06}
elif salary <= 15:
    inc = {'A':0.08,'B':0.06,'C':0.04,'D':0}
else:
    inc = {'A':0.07,'B':0.05,'C':0.04,'D':0}

print("Increment:", salary * inc[rating])
```

```python
from datetime import datetime

start_date = input("Enter FD start date (dd-mm-yyyy): ")
end_date = input("Enter FD end date (dd-mm-yyyy): ")

s_date = datetime.strptime(start_date, "%d-%m-%Y")
e_date = datetime.strptime(end_date, "%d-%m-%Y")

days = (e_date - s_date).days

if days <= 180:
    print("Interest Rate: 4%")
elif days <= 365:
    print("Interest Rate: 5.5%")
elif days <= 1095:
    print("Interest Rate: 6.5%")
else:
    print("Interest Rate: 7%")
```

```python
#Name validation
def validate_name(name):
    if name.isalpha():
        print("Valid name")
    else:
        print("Invalid name")

validate_name(input("Enter name: "))
```

```python
#Mobile number validation

def validate_mobile(m):
    if m.isdigit() and len(m) == 10:
        print("Valid mobile")
    else:
        print("Invalid mobile")

validate_mobile(input("Enter mobile: "))
```

```python
#Customer dictionary

cust = {}
cid = 1

while True:
    name = input("Name: ")
    dob = input("DOB: ")
    mobile = input("Mobile: ")

    cust[cid] = {"name": name, "DOB": dob, "mobile": mobile}
    cid += 1

    if input("Add more? y/n: ") == 'n':
        break

print(cust)
```

```python
class User:
    def signup(self):
        print("User signed up")

    def signin(self):
        print("User signed in")

    def view_products(self):
        print("Viewing products")

    def place_order(self):
        print("Order placed")

u = User()
u.signup()
u.signin()
u.view_products()
u.place_order()
```
