```python
import sqlite3
```

```python
conn = sqlite3.connect("hotel.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS hotel_bill (
    item_name TEXT,
    quantity INTEGER,
    price INTEGER,
    total INTEGER
)
""")
conn.commit()
```

```python
menu = {
    "Paneer Butter Masala": 250,
    "Dal Fry": 180,
    "Butter Naan": 40,
    "Jeera Rice": 150,
    "Noodles":210,
    "Pizza":200,
    "Samosa": 30
}
```

```python
print("Welcome to Bajrangbali Da Dhaba\n")
print("Menu:")

for item, price in menu.items():
    print(item, "₹", price)

item_name = input("\nEnter item name: ")
quantity = int(input("Enter quantity: "))
```

```python
price = menu[item_name]
total = price * quantity
```

```python
cursor.execute(
    "INSERT INTO hotel_bill VALUES (?, ?, ?, ?)",
    (item_name, quantity, price, total)
)
conn.commit()
```

```python
print("\n----------- HOTEL BILL -----------")
print("Restaurant : Bajrangbali Da Dhaba")
print("Item       :", item_name)
print("Quantity   :", quantity)
print("Price      : ₹", price)
print("Total Amt  : ₹", total)
print("---------------------------------")
```

```python
print("\nSaved Data in Database:")
cursor.execute("SELECT * FROM hotel_bill")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
conn.close()
```
